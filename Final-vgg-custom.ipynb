## Building the VGG Model for Hyper Parameter Tuning with Keras Tuner

```python
def build_model_hyper_param(hp):
        """Builds a model that follows the VGG architecture with Keras Tuner to determine the best model"""
        model = tf.keras.models.Sequential()

        model.add(tf.keras.layers.Conv2D(input_shape=(28,28,3), filters=64, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))
        model.add(tf.keras.layers.Dropout(0.25))

        model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))
        model.add(tf.keras.layers.Dropout(0.25))

        model.add(tf.keras.layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", activation="relu"))
        model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2)))

        model.add(tf.keras.layers.Flatten())
        
        
        hp_unit1 = hp.Choice('units_1', values=[32, 64, 128, 256])
        model.add(tf.keras.layers.Dense(hp_unit1))


        hp_unit2 = hp.Choice('dropout_1', values=[0.05, 0.10, 0.20, 
                                                0.25, 0.30,0.35])    
        model.add(tf.keras.layers.Dropout(hp_unit2))


        hp_unit3 = hp.Choice('units_3', values=[16, 32, 64])   
        model.add(tf.keras.layers.Dense(hp_unit3)) 


        hp_unit4 = hp.Choice('dropout_2', values=[0.05, 0.10, 0.20,
                                                0.25, 0.30,0.35])
        model.add(tf.keras.layers.Dropout(hp_unit4))

        model.add(tf.keras.layers.Dense(10, activation="softmax"))


        hp_learning_rate = hp.Choice('learning_rate',
                                     values=[1e-3, 3e-3, 1e-4, 3e-4]) 

        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate ), 
                      loss=tf.keras.losses.CategoricalCrossentropy(),
                      metrics="acc")
        return model

```

### Build Callbacks for Hyper Parameter Tuning

```python
# Build Callbacks
early_stopper = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=5
)
```

### Perform Hyper Parameter Tuning with Bayesian Optimization

```python
tuner = kt.BayesianOptimization(
    build_model_hyper_param,
    objective=kt.Objective("val_loss", "min"),
    max_trials=25,
    directory="results",
    project_name="final_vgg_hyper_training",
    overwrite=False,
)
```

```python
tuner.search(train_dataset,
             epochs=1000,
             validation_data=(val_images, val_labels_oh), 
             callbacks=[early_stopper] 
            )
```

```python
tuner.results_summary()
```

```python
# save tuner object
with open("results/vgg_tuner.pkl", "wb") as f:
    pickle.dump(tuner, f)
```

## Building the VGG Custom Model with the Best Hyper Parameters

```python
def get_best_vgg_model():
    
        model = tf.keras.models.Sequential()

        model.add(tf.keras.layers.Conv2D(input_shape=(28,28,3), filters=64, kernel_size=(3,3), 
                                         padding="same", activation="relu", name="block1_conv1"))
        model.add(tf.keras.layers.Conv2D(filters=64, kernel_size=(3,3), padding="same",
                                         activation="relu", name="block1_conv2"))
        model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2), name="block1_pool"))
        model.add(tf.keras.layers.Dropout(0.25, name="block1_dropout"))

        model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", 
                                         activation="relu", name="block2_conv1"))
        model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", 
                                         activation="relu", name="block2_conv2"))
        model.add(tf.keras.layers.Conv2D(filters=128, kernel_size=(3,3), padding="same", 
                                         activation="relu", name="block2_conv3"))
        model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2), name="block2_pool"))
        model.add(tf.keras.layers.Dropout(0.25, name="block2_dropout"))

        model.add(tf.keras.layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", 
                                         activation="relu", name="block3_conv1"))
        model.add(tf.keras.layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", 
                                         activation="relu", name="block3_conv2"))
        model.add(tf.keras.layers.Conv2D(filters=256, kernel_size=(3,3), padding="same", 
                                         activation="relu", name="block3_conv3"))
        model.add(tf.keras.layers.MaxPool2D(pool_size=(2,2),strides=(2,2), name="block3_pool"))

        model.add(tf.keras.layers.Flatten())
        
        model.add(tf.keras.layers.Dense(256, name="dense_1"))
        model.add(tf.keras.layers.Dropout(0.05, name="dense_dropout_1"))
        model.add(tf.keras.layers.Dense(16, name="dense_2")) 
        model.add(tf.keras.layers.Dropout(0.35, name="dense_dropout_2"))

        model.add(tf.keras.layers.Dense(10, activation="softmax", name="dense_output"))
        
        model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001 ),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics="acc")

        return model
```

```python
model = get_best_vgg_model()
```

```python
model.summary()
```

### Build Callbacks for the Final Model Training

```python
# Build Callbacks
model_checkpointer = tf.keras.callbacks.ModelCheckpoint(
    "models/final/final_vgg",
    monitor = "val_loss",
    save_best_only = True,
    save_weights_only = True
)

early_stopper = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=5
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                              patience=5, min_lr=0.000001, verbose=1)
```

## Model Training

```python
# If kernel crashes or you get an resource exhauseted error, reduce batch size, always use powers of 2 eg: 32, 62, 128...
history = model.fit(train_dataset, epochs=1000, validation_data=(val_images, val_labels_oh), 
                    callbacks=[model_checkpointer, early_stopper, reduce_lr])
```
