## Load EfficientNetB1 Model Trained on ImageNet Data

```python
# Load the pre trained model
# Top is not included for us to customize
base_model = efn.EfficientNetB1(weights='imagenet', include_top=False)
```

## Building the Efficientnet B1 Model for Hyper Parameter Tuning

```python
def build_model_hyper_param(hp):
    """Builds a model that utilizes transfer learning from the pre trained efficientnet b1 with Keras Tuner to determine the best model"""
    inputs = tf.keras.layers.Input(shape=(28, 28, 3))

    x = base_model(inputs)
    x = tf.keras.layers.Flatten()(x)    

    hp_unit1 = hp.Choice('units_1', values=[32, 64, 128,])
    x = tf.keras.layers.Dense(hp_unit1)(x)


    hp_unit2 = hp.Choice('dropout_1', values=[0.05, 0.10, 0.20, 
                                        0.25, 0.30,0.35])    
    x = tf.keras.layers.Dropout(hp_unit2)(x)


    hp_unit3 = hp.Choice('units_3', values=[16, 32, 64])   
    x = tf.keras.layers.Dense(hp_unit3)(x)


    hp_unit4 = hp.Choice('dropout_2', values=[0.05, 0.10, 0.20,
                                        0.25, 0.30,0.35])

    x = tf.keras.layers.Dropout(hp_unit4)(x)

    outputs = tf.keras.layers.Dense(10, activation="softmax")(x)


    hp_learning_rate = hp.Choice('learning_rate',
                             values=[1e-3, 3e-3, 1e-4, 3e-4]) 

    model = tf.keras.Model(inputs=inputs, outputs=outputs)

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=hp_learning_rate ), 
                loss=tf.keras.losses.CategoricalCrossentropy(),
                  metrics="acc")
    
    return model
```

### Build Callbacks for Hyper Parameter Tuning

```python
# Build Callbacks
early_stopper = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=3
)
```

### Perform Hyper Parameter Tuning with Bayesian Optimization

```python
tuner = kt.BayesianOptimization(
    build_model_hyper_param,
    objective=kt.Objective("val_loss", "min"),
    max_trials=60,
    directory="results",
    project_name="final_efficientnetb1_hyper_training",
    overwrite=False,
)
```

```python
tuner.search(train_dataset,
             epochs=1000,
             validation_data=(val_images, val_labels_oh), 
             callbacks=[early_stopper] 
            )
```

## Building the Efficienetnet Model with the Best Hyper Parameters and SAM (Sharpness-Aware Minimization)

Sharpness-Aware minimization (SAM) is a technique that improves the model generalization and provides robustness to label noise

```python
def get_best_eff_model():
    """
    Model with the best Hyper parameters
    """
    
    inputs = tf.keras.layers.Input(shape=(28, 28, 3))

    x = base_model(inputs)
    x = tf.keras.layers.Flatten()(x)    

    x = tf.keras.layers.Dense(32)(x)
    x = tf.keras.layers.Dropout(0.25)(x)
    x = tf.keras.layers.Dense(32)(x)
    x = tf.keras.layers.Dropout(0.05)(x)

    outputs = tf.keras.layers.Dense(10, activation="softmax")(x)

    model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return model
```

```python
eff_model = get_best_eff_model()
```

```python
class SAMEffModel(tf.keras.Model):
    """
    Build a Model Subcalss which utlizes the effientnet model along with Sharpness-Aware Minimization
    """
    def __init__(self, model, rho=0.05):
        super(SAMEffModel, self).__init__()
        self.model = (model)
        self.rho = rho

    def train_step(self, data):
        (images, labels) = data
        e_ws = []
        with tf.GradientTape() as tape:
            predictions = self.model(images)
            loss = self.compiled_loss(labels, predictions)
        trainable_params = self.model.trainable_variables
        gradients = tape.gradient(loss, trainable_params)
        grad_norm = self._grad_norm(gradients)
        scale = self.rho / (grad_norm + 1e-12)

        for (grad, param) in zip(gradients, trainable_params):
            e_w = grad * scale
            param.assign_add(e_w)
            e_ws.append(e_w)

        with tf.GradientTape() as tape:
            predictions = self.model(images)
            loss = self.compiled_loss(labels, predictions)    
        
        sam_gradients = tape.gradient(loss, trainable_params)
        for (param, e_w) in zip(trainable_params, e_ws):
            param.assign_sub(e_w)
        
        self.optimizer.apply_gradients(
            zip(sam_gradients, trainable_params))
        
        self.compiled_metrics.update_state(labels, predictions)
        return {m.name: m.result() for m in self.metrics}
    
    def call(self, x):
        return self.model(x)

    def test_step(self, data):
        (images, labels) = data
        predictions = self.model(images, training=False)
        loss = self.compiled_loss(labels, predictions)
        self.compiled_metrics.update_state(labels, predictions)
        return {m.name: m.result() for m in self.metrics}

    def _grad_norm(self, gradients):
        norm = tf.norm(
            tf.stack([
                tf.norm(grad) for grad in gradients if grad is not None
            ])
        )
        return norm
```

```python
# Best efficentnet model with SAM
model = SAMEffModel(eff_model)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001 ), 
            loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics="acc")
```

### Build Callbacks for Final Model Tuning

```python
# Build Callbacks
model_checkpointer = tf.keras.callbacks.ModelCheckpoint(
    "models/final/final_eff",
    monitor = "val_loss",
    save_best_only = True,
    save_weights_only = True
)

early_stopper = tf.keras.callbacks.EarlyStopping(
    monitor='val_loss',
    min_delta=0,
    patience=3
)

reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.3,
                              patience=5, min_lr=0.000001, verbose=1)
```

## Model Training

```python
# If kernel crashes or you get an resource exhauseted error, reduce batch size, always use powers of 2 eg: 32, 62, 128...
history = model.fit(train_dataset, epochs=1000, validation_data=(val_images, val_labels_oh), 
                    callbacks=[model_checkpointer, early_stopper, reduce_lr])
```

## Training Assesment

```python
def plot_history(history):
    loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' not in s]
    val_loss_list = [s for s in history.history.keys() if 'loss' in s and 'val' in s]
    acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' not in s]
    val_acc_list = [s for s in history.history.keys() if 'acc' in s and 'val' in s]
    
    if len(loss_list) == 0:
        print('Loss is missing in history')
        return 
    
    ## As loss always exists
    epochs = range(1,len(history.history[loss_list[0]]) + 1)
    
    ## Loss
    plt.figure(1)
    for l in loss_list:
        plt.plot(epochs, history.history[l], 'b', label='Training loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    for l in val_loss_list:
        plt.plot(epochs, history.history[l], 'g', label='Validation loss (' + str(str(format(history.history[l][-1],'.5f'))+')'))
    
    plt.title('Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    
    ## Accuracy
    plt.figure(2)
    for l in acc_list:
        plt.plot(epochs, history.history[l], 'b', label='Training accuracy (' + str(format(history.history[l][-1],'.5f'))+')')
    for l in val_acc_list:    
        plt.plot(epochs, history.history[l], 'g', label='Validation accuracy (' + str(format(history.history[l][-1],'.5f'))+')')

    plt.title('Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.show()
```

```python
# Plot Loss and Accuracy of Training History
plot_history(history)
```

**Eventhough the model seems to underfit, it is due to the MixUp Data augmentation.**

**In the evaluation we see the train accuracy is in fact higher than the validation accuracy, therefore the model has perfectly fitted.**
